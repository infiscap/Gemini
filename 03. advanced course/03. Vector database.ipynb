## 1.1 정의

벡터 데이터베이스 는 벡터 임베딩 또는 텍스트, 이미지, 오디오와 같은 구조화되지 않은 데이터의 수치 표현을 저장, 색인 생성, 쿼리할 수 있게 해주는 모든 데이터베이스
