```python
PROJECT_ID=!gcloud config get-value project
PROJECT_ID=PROJECT_ID[0]
```

```python
from google.cloud import translate_v3 as translate
```

```python
client = translate.TranslationServiceClient()
parent = client.common_project_path(PROJECT_ID)
display_language_code = 'ko'

response = client.get_supported_languages(
    parent=parent,
    display_language_code=display_language_code
)
```

```python
print(response)
```

```python
languages = response.languages

print(f" Languages: {len(languages)}".center(60, "-"))
for language in languages:
    lang_code = language.language_code
    lang_name = language.display_name
    print(f"{lang_code:10} : {lang_name}")
```

```python
display_language_code = 'ja'

response = client.get_supported_languages(
    parent=parent,
    display_language_code=display_language_code
)

languages = response.languages

print(f" Languages: {len(languages)}".center(60, "-"))
for language in languages:
    lang_code = language.language_code
    lang_name = language.display_name
    print(f"{lang_code:10} : {lang_name}")
```

```python
txt="길마가 용계 입구로 모이시랍니다."
target_leng="en"

response = client.translate_text(
      parent=parent,
      contents=[txt],
      target_language_code=target_leng
)

print(response)
print(response.translations)
```

```python

target_languages = ["tr", "de", "es", "it", "el", "zh", "ja", "en"]

print(f" {txt} ".center(50, "-"))

for target_lang in target_languages:
  response = client.translate_text(
      parent=parent,
      contents=[txt],
      target_language_code=target_lang
  )

  translation = response.translations[0]
  print(f" {translation.detected_language_code:3} -> {target_lang:3} : {translation.translated_text}")
```

```python
sentences=[]

for target_lang in target_languages:
  response = client.translate_text(
      parent=parent,
      contents=[txt],
      target_language_code=target_lang
  )

  sentences.append(response.translations[0].translated_text)
```

```python
for sentence in sentences:
  response = client.detect_language(
      parent=parent,
      content=sentence
  )
  lang = response.languages[0]

  print(f"Confidence: {lang.confidence:4.0%}",
        f"Language: {lang.language_code:5}",
        sentence,
        sep=" | ")
```
