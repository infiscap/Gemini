```
# agent.py 파일
import datetime # 날짜 및 시간 관련 기능을 제공하는 파이썬 내장 모듈을 임포트합니다.
from zoneinfo import ZoneInfo # 특정 시간대 정보를 다루기 위한 모듈을 임포트합니다.
from google.adk.agents import Agent # ADK에서 에이전트를 정의하고 생성하는 데 사용되는 Agent 클래스를 임포트합니다.
# 날씨 정보를 가져오는 함수
# Function to get weather information
def get_weather(city: str) -> dict:
    """특정 도시의 현재 날씨 보고서를 검색합니다.
    Args:
        city (str): 날씨 보고서를 검색할 도시의 이름입니다.
    Returns:
        dict: 상태 및 결과 또는 오류 메시지를 담고 있는 딕셔너리입니다.
              예: {"status": "success", "report": "..."} 또는 {"status": "error", "error_message": "..."}
    """
    if city.lower() == "new york": # 입력된 도시 이름을 소문자로 변환하여 "new york"과 비교합니다.
        return {
            "status": "success", # 요청이 성공했음을 나타내는 상태 코드입니다.
            "report": ( # 날씨 보고서 내용을 담고 있는 문자열입니다.
                "The weather in New York is sunny with a temperature of 25 degrees"
                " Celsius (77 degrees Fahrenheit)."
            ),
        }
    else: # 입력된 도시가 "new york"이 아닌 경우.
        return {
            "status": "error", # 요청이 실패했음을 나타내는 상태 코드입니다.
            "error_message": f"Weather information for '{city}' is not available.", # 특정 도시에 대한 날씨 정보가 없다는 오류 메시지입니다.
        }
# 현재 시간을 가져오는 함수
# Function to get current time
def get_current_time(city: str) -> dict:
    """특정 도시의 현재 시간을 반환합니다.
    Args:
        city (str): 현재 시간을 검색할 도시의 이름입니다.
    Returns:
        dict: 상태 및 결과 또는 오류 메시지를 담고 있는 딕셔너리입니다.
              예: {"status": "success", "report": "..."} 또는 {"status": "error", "error_message": "..."}
    """
    if city.lower() == "new york": # 입력된 도시 이름을 소문자로 변환하여 "new york"과 비교합니다.
        tz_identifier = "America/New_York" # 뉴욕에 대한 시간대 식별자를 설정합니다.
    else: # 입력된 도시가 "new york"이 아닌 경우.
        return {
            "status": "error", # 요청이 실패했음을 나타내는 상태 코드입니다.
            "error_message": ( # 특정 도시에 대한 시간대 정보가 없다는 오류 메시지입니다.
                f"Sorry, I don't have timezone information for {city}."
            ),
        }
    tz = ZoneInfo(tz_identifier) # 설정된 시간대 식별자를 사용하여 ZoneInfo 객체를 생성합니다.
    now = datetime.datetime.now(tz) # 현재 시간을 특정 시간대 기준으로 얻습니다.
    report = ( # 포맷팅된 현재 시간 정보를 담고 있는 문자열입니다.
        f'The current time in {city} is {now.strftime("%Y-%m-%d %H:%M:%S %Z%z")}'
    )
    return {"status": "success", "report": report} # 성공 상태와 포맷팅된 시간 정보를 반환합니다.
# 에이전트 인스턴스 생성
# Create agent instance
root_agent = Agent(
    name="weather_time_agent", # 에이전트의 고유한 이름입니다.
    model="gemini-2.0-flash", # 에이전트가 사용할 LLM 모델을 지정합니다. Gemini API를 사용할 수 있도록 설정하는 방법에 대해서는 이전에 논의했습니다.
    description=( # 에이전트에 대한 간략한 설명입니다.
        "Agent to answer questions about the time and weather in a city."
    ),
    instruction=( # 에이전트의 역할과 수행해야 할 작업에 대한 지침입니다.
        "You are a helpful agent who can answer user questions about the time and weather in a city."
    ),
    tools=[get_weather, get_current_time], # 에이전트가 사용할 수 있는 도구 (함수) 목록입니다. 여기서는 앞에서 정의한 get_weather와 get_current_time 함수를 등록합니다.
)
위 주석을 통해 agent.py 파일의 각 부분이 어떤 역할을 하는지 더 명확하게 이해할 수 있을 것입니다. 이 파일은 기본적인 날씨 및 시간 정보를 제공하는 에이전트를 정의하고 있습니다.

```
