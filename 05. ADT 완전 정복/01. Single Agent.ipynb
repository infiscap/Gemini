# 1. 이해하기

* **AI 에이전트 개발**: ADK는 **AI 에이전트를 개발**하기 위한 **프레임워크**입니다. 개발자들이 AI를 활용하여 사용자를 대신해 목표를 추구하고 작업을 완료하는 소프트웨어 시스템을 구축할 수 있도록 지원합니다.
* **오픈소스**: ADK는 **Google에서 개발한 오픈소스 Python 툴킷**입니다. Apache 2.0 라이선스 하에 공개 및 배포되고 있습니다.
* **멀티 에이전트 시스템**: ADK는 **여러 개의 독립적인 에이전트를 조합**하여 멀티 에이전트 시스템을 구축하는 데 용이합니다. 이를 통해 **복잡한 워크플로우를 처리**하고, **에이전트 간의 협업을 가능**하게 합니다.
* **Gemini 및 Google 생태계 최적화**: ADK는 Gemini 모델과 Google Cloud 플랫폼에 최적화되어 있지만, **모델이나 배포 방식에 구애받지 않고 다른 프레임워크와의 호환성**도 고려하여 설계되었습니다.
* **코드 중심 개발**: ADK는 GUI 기반 설정 대신 **Python 코드**를 통해 **에이전트의 동작, 오케스트레이션 및 도구 사용을 직접 정의하는 코드 중심(code-first) 개발 방식**을 채택하고 있습니다. 이를 통해 개발자는 에이전트에 대한 유연하고 세밀한 제어를 할 수 있으며, 테스트 가능성을 높입니다.
